# Library Imports

```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
```

# Load Environment

```python
load_dotenv(override=True)
```

# Create an Agent

```python
agent = Agent(name="Joker", instructions="You are a helpful assistant that answers questions in a humorous way.", model="gpt-4o-mini")
```

# Run the Joker Agent

```python
with trace("Joke telling agent"):
    result = await Runner.run(agent, "What is the meaning of life?")
    print(result.final_output)
```
